```python
class DataPipeline():
    def __init__(self, dataset):
        """Constructor for DataPipeline class

        initial_data: copy of the desired datset
        replace_val: Value to replace with nan values or missing values
        ordinal_encoder: OrdinalEncoder object of Scikit-Learn to encode columns
        valid_columns: Columns to take in account
        """
        self.original_data = dataset.copy()
        self.initial_data = dataset.copy()
        self.replace_val = -1
        self.ordinal_encoder = OrdinalEncoder(dtype = np.int32, handle_unknown = "use_encoded_value", unknown_value = self.replace_val)
        self.valid_columns = ["Yas", "Cinsiyet", "Uyruk", "Bolum"]

    def __fillAbsentValues(self,):
        """Find proper values for each HastaNo and replace missing values with corresponding ids

        """

        id_col = "HastaNo"
        target = "UygulamaSüresi"
        columns = dataset.columns

        # For each column
        for col in columns:
          if col == target:
            return
          absent_ids = self.initial_data[self.initial_data[col].isnull()][id_col].value_counts().keys()

          # Find each matched value for hastano to replace with nan values
          ids_with_proper_values = {}
          for id in absent_ids:
            patient_data = self.initial_data[self.initial_data[id_col] == id][col]
            non_absent_entries = patient_data.dropna().unique()

            if len(non_absent_entries) > 0:
              ids_with_proper_values[id] = non_absent_entries[0]

          # Fill the missing values with found value
          for id, value in ids_with_proper_values.items():
            mask = (self.initial_data[id_col] == id) & (self.initial_data[col].isna())
            self.initial_data.loc[mask, col] = value

    def __fillNaValues(self,):
        """Fill missing values with replace_val

        """

        self.initial_data.fillna(self.replace_val, inplace = True)
        self.initial_data.replace("nan", self.replace_val, inplace = True)
        self.initial_data.replace("NAN", self.replace_val, inplace = True)
        self.initial_data.replace(np.nan, self.replace_val, inplace = True)

    def __separateBloodRh(self, ):
        """Creates two columns as "Sadece_Kan_Grubu"(A, 0, AB, B) and "Sadece_Rh_Tipi"(+, -)
        """

        self.initial_data["KanGrubu"] = self.initial_data["KanGrubu"].astype(str)
        self.initial_data["Sadece_Kan_Grubu"] = self.initial_data["KanGrubu"].apply(lambda x: x.split(" ")[0])
        self.initial_data["Sadece_Rh_Tipi"] = self.initial_data["KanGrubu"].apply(lambda x: x.split(" ")[-1][-1])
        self.valid_columns.append("Sadece_Kan_Grubu")
        self.valid_columns.append("Sadece_Rh_Tipi")

    def __separateAllergy(self,):
        """For each allergy value present in the "Alerji" column create additional other columns to reduce number of values inside a column
        """

        self.initial_data["Alerji"] = self.initial_data["Alerji"].astype(str)
        self.initial_data["Alerji"] = self.initial_data["Alerji"].apply(lambda x: x.strip().lower())

        def getRelevantAllergy(text, index):
            curr_array = text.split(",")
            if index >= len(curr_array):
              return self.replace_val

            return curr_array[index]

        iteration_size = self.initial_data["Alerji"].apply(lambda x: len(x.split(","))).values.max()

        for i in range(iteration_size):
            self.initial_data[f"Alerji_{i + 1}"] = self.initial_data["Alerji"].apply(lambda x: getRelevantAllergy(x, i))
            self.valid_columns.append(f"Alerji_{i + 1}")

    def __timeStandardization(self,):
        """Standardize time related values and creates two new columns for them
        """

        self.initial_data["Tedavi_Süresi_Seans"] = self.initial_data["TedaviSuresi"].apply(lambda x: x.split(" ")[0])
        self.initial_data["Uygulama_Süresi_Dakika"] = self.initial_data["UygulamaSuresi"].apply(lambda x: x.split(" ")[0])
        self.valid_columns.append("Tedavi_Süresi_Seans")
        self.valid_columns.append("Uygulama_Süresi_Dakika")

    def __separateChronicalSickness(self, ):
        """For each chronical sickness value present in the "KronikHastalik" column create additional other columns to reduce number of values inside a column
        """

        self.initial_data["KronikHastalik"] = self.initial_data["KronikHastalik"].astype(str)
        self.initial_data["KronikHastalik"] = self.initial_data["KronikHastalik"].apply(lambda x: x.strip().lower())

        def getRelevantChronicalSickness(text, index):
            curr_array = text.split(",")
            if index >= len(curr_array):
              return self.replace_val

            return curr_array[index]

        iteration_size = self.initial_data["KronikHastalik"].apply(lambda x: len(x.split(","))).values.max()

        for i in range(iteration_size):
            self.initial_data[f"Kronik_Hastalik_{i + 1}"] = self.initial_data["KronikHastalik"].apply(lambda x: getRelevantChronicalSickness(x, i))
            self.valid_columns.append(f"Kronik_Hastalik_{i + 1}")

    def __encodeColumns(self,):
        """Apply ordinal_encoder object here to encode relevant columns
        """
        columns_to_encode = [col for col in self.valid_columns if col not in ["Tedavi_Süresi_Seans", "Uygulama_Süresi_Dakika", "Yas"]]
        for col in columns_to_encode:
            self.initial_data[col] = self.initial_data[col].astype(str)
        self.initial_data[columns_to_encode] = self.ordinal_encoder.fit_transform(self.initial_data[columns_to_encode])

    def __logTarget(self,):
      """Apply np.log1p (Natural Logarithm) to the target for normalization
      """
      self.initial_data["Uygulama_Süresi_Dakika"] = self.initial_data["Uygulama_Süresi_Dakika"].astype(float)
      self.initial_data["Uygulama_Süresi_Dakika"] = np.log1p(self.initial_data["Uygulama_Süresi_Dakika"])
      print("TARGET COLUMN HAS NORMALIZED WITH LOG1P")

    def process_data(self, apply_log = False):
        """Create dataset
        """

        self.__fillAbsentValues()
        self.__fillNaValues()
        self.__timeStandardization()
        self.__separateBloodRh()
        self.__separateAllergy()
        self.__separateChronicalSickness()
        if apply_log:
          self.__logTarget()
        self.__encodeColumns()
        return self.initial_data[self.valid_columns]

    def __call__(self, apply_log = False):
        """Same as calling setup but defines new syntax and make it more understandable and cleaner
        """
        return self.process_data(apply_log)
```
